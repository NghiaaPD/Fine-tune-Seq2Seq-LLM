```python
import pandas as pd

import matplotlib.pyplot as plt

# c file kt qu
df = pd.read_csv('results.txt', sep='\t', comment='/', engine='python')

# V cc ch s chnh theo epoch
metrics = ['train_loss', 'valid_loss', 'rouge1', 'rouge2', 'rougeL', 'bertscore_f1', 'bleu', 'meteor', 'sacrebleu']
plt.figure(figsize=(18, 12))

for i, metric in enumerate(metrics, 1):
    plt.subplot(3, 3, i)
    plt.plot(df['epoch'], df[metric], marker='o')
    plt.title(metric)
    plt.xlabel('Epoch')
    plt.ylabel(metric)
    plt.grid(True)

plt.tight_layout()
plt.show()
```
